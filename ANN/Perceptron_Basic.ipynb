```python
import numpy as np

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Make a Dataset with two input columns and one output column


```python
from sklearn.datasets import make_blobs

X,y=make_blobs(n_samples=200, centers=2, random_state=42)
```

```python
X.shape
```

```python
y
```

```python
X[:,1]
```

```python
sns.scatterplot( x=X[:,0], y=X[:,1] ,hue=y)
```

```python
from sklearn.linear_model import Perceptron

per = Perceptron()

per.fit(X,y)
```

```python
per.coef_       # weights of the perceptron after training 
```

```python
per.intercept_   # Bias of the perceptron after training 
```
