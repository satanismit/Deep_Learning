```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img( '/content/frog.jpeg', target_size=(200,200))
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
img = image.img_to_array(img)

img.shape
```

```python
# create an object
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect',

)
```

```python
# we need to convert input as batch

input_batch = img.reshape(1,200,200,3)
```

```python
# creating folder
import os

save_dir = '/content/augmented_images'
os.makedirs(save_dir, exist_ok=True)
```

```python
# now generate 10 images from one image
i=0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir=save_dir, save_prefix='frog', save_format='jpeg'):
  i=i+1


  if i==10:
    break;
```
